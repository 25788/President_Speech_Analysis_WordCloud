Ayrıca negative_prompt ile resimde olmasını istemediğimiz şeyleri belirtiyoruz. controlnet_conditioning_scale ile yüklediğimiz resmin yeni oluşan resimdeki etkisini ayarlıyoruz. Bu değişken ne kadar yüksek olursa yüklenen resmin etkisi o kadar çok olur. num_inference_steps ile de resmin kaç adımda oluşturulacağını belirliyoruz.
